```python
df_clean = df.copy()
df_clean
```

```python
df_clean.dropna(subset=['Place','Area','Const Year','Price'],inplace=True) #added inplace=True as the data was not dropping Nan
df_clean
```

```python
df_clean.isna().sum()
```

```python
df_clean['Const Year'] = df_clean['Const Year'].astype(int)
#df_clean["Const Year"] = pd.to_numeric(df_clean["Const Year"], errors="coerce").astype(int)
```
